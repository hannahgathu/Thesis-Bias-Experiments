```python
import seaborn as sb
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
def read_file(url):

    url = url + "?raw=true"
    df = pd.read_csv(url, encoding='cp1252', sep=',', low_memory=False)
    return df
```

```python
url = "https://github.com/hannahgathu/Thesis-Data-Visualisations/blob/main/All_Data_Prepped.csv"

df= read_file(url).iloc[:65378,:]
```

```python
features= df.iloc[:,3:]
```

```python
# scaler = StandardScaler()
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

## Covariance Matrix

```python
scaled_features_transposed=np.transpose(scaled_features)
```

```python
covMatrix = np.cov(scaled_features_transposed, bias=True)
plt.figure(figsize = (25,15))
sb.heatmap(covMatrix, annot=True, fmt='g')
```

```python
covMatrix_rounded = np.around(covMatrix,decimals=3)
plt.figure(figsize = (25,15))
sb.heatmap(covMatrix_rounded, annot=True, fmt='g')
```

## Correlation Matrix

```python
corrMatrix = np.corrcoef(scaled_features_transposed)
```

```python
plt.figure(figsize = (25,15))
sb.heatmap(corrMatrix,annot = True)
```

```python
corrMatrix_rounded = np.around(corrMatrix,decimals=3)
plt.figure(figsize = (25,15))
sb.heatmap(corrMatrix_rounded, annot=True, fmt='g')
```
