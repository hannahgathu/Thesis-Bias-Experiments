```python
sum = 0.9907414834431378
#sum = 1.0010752688172042
#a=1/93
#a=1/45
a=1/28
#k=0.1
#k=0.05
k=0
def f(x):
    f = a*(x-4)**2+k/sum
    return f
    
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1 = [0, 1, 2, 3, 4, 5, 6] #left edge
y = [f(1),f(2),f(3),f(4),f(5),f(6),f(7)]
#w = np.array(x2) - np.array(x1) #variable width, can set this as a scalar also
plt.bar(x1, y)
plt.ylim(0, 0.3)
plt.show()
```
