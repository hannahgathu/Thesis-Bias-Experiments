```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
metrics_tables_baseline_dict={}
#for i in range(10):
for i in [0,1,2,3,4,5,6,9]:
    metrics_tables_baseline_dict["{}".format(i)] = pd.read_csv("./baseline/metrics_table_{}.csv".format(i))
```

```python
metrics_tables_density_dict={}
for i in range(10):
    print(i)
    metrics_tables_density_dict["{}".format(i)] = pd.read_csv("./density/metrics_table_{}.csv".format(i))
```

```python
metrics_tables_uncertainty_dict={}
for i in range(10):
    metrics_tables_uncertainty_dict["{}".format(i)] = pd.read_csv("./uncertainty/metrics_table_{}.csv".format(i),  error_bad_lines=False)
```

```python
metrics_table_density=pd.concat(metrics_tables_density_dict.values())
```

```python
metrics_table_baseline=pd.concat(metrics_tables_baseline_dict.values())
```

```python
metrics_table_uncertainty=pd.concat(metrics_tables_uncertainty_dict.values())
```

```python
metrics_table_baseline.shape[0]
```

```python
metrics_table_baseline['index']=metrics_table_baseline.iloc[:,0]
```

```python
metrics_table_density['index']=metrics_table_density.iloc[:,0]
```

```python
metrics_table_uncertainty['index']=metrics_table_uncertainty.iloc[:,0]
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
p=sns.lineplot(x = "index", y = "accuracy", ci="sd", data = metrics_table_uncertainty)
p.set(xlim=(0,4000),ylim=(0.7,1))
p.axes.set_title("accuracy",fontsize=30)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
p=sns.lineplot(x = "index", y = "accuracy", ci="sd", data = metrics_table_uncertainty, label="uncertainty only")
k=sns.lineplot(data=metrics_table_density, x = "index", y = "accuracy", ci="sd", label="with density")
#m=sns.lineplot(data=metrics_table_baseline, x = "index", y = "accuracy", ci="sd", label="baseline")
p.set(xlim=(0,2500),ylim=(0.9,1))
p.axes.set_title("accuracy",fontsize=30)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
p=sns.lineplot(x = "index", y = "balanced_accuracy", ci="sd", data = metrics_table_uncertainty, label="uncertainty only")
k=sns.lineplot(data=metrics_table_density, x = "index", y = "balanced_accuracy", ci="sd", label="with density")
#m=sns.lineplot(data=metrics_table_baseline, x = "index", y = "accuracy", ci="sd", label="baseline")
p.set(xlim=(0,2500),ylim=(0.5,1))
p.axes.set_title("balanced accuracy",fontsize=30)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
p=sns.lineplot(x = "index", y = "pr_auc", ci="sd", data = metrics_table_uncertainty, label="uncertainty only")
k=sns.lineplot(data=metrics_table_density, x = "index", y = "pr_auc", ci="sd", label="with density")
#m=sns.lineplot(data=metrics_table_baseline, x = "index", y = "accuracy", ci="sd", label="baseline")
p.set(xlim=(0,2000),ylim=(0.6,1))
p.axes.set_title("PR AUC",fontsize=30)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
p=sns.lineplot(x = "index", y = "recall", ci="sd", data = metrics_table_uncertainty, label="uncertainty only")
k=sns.lineplot(data=metrics_table_density, x = "index", y = "recall", ci="sd",label="with density")
#m=sns.lineplot(data=metrics_table_baseline, x = "index", y = "recall", ci="sd", label="baseline")
p.set(xlim=(0,100),ylim=(0.4,0.95))
p.axes.set_title("recall",fontsize=30)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
p=sns.lineplot(x = "index", y = "f2_score", ci="sd", data = metrics_table_uncertainty, label="uncertainty only")
k=sns.lineplot(data=metrics_table_density, x = "index", y = "f2_score", ci="sd",label="with density")
#m=sns.lineplot(data=metrics_table_baseline, x = "index", y = "f2_score", ci="sd", label="baseline")
p.set(xlim=(0,2500),ylim=(0.45,0.95))
p.axes.set_title("f2_score",fontsize=30)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
p=sns.lineplot(x = "index", y = "precision", ci="sd", data = metrics_table_uncertainty, label="uncertainty only")
k=sns.lineplot(data=metrics_table_density, x = "index", y = "precision", ci="sd",label="with density")
#m=sns.lineplot(data=metrics_table_baseline, x = "index", y = "precision", ci="sd", label="baseline")
p.set(xlim=(0,2500),ylim=(0.5,1))
p.axes.set_title("Precision",fontsize=30)
```
